```python
# 선택정렬 (Selection Sorting)

# 내 코드
def selection_sort(array):
    i = 0
    while i != len(array)-1:
        min_idx= array.index(min(array[i:]))
        if array[min_idx] <= array[i]:
            array[min_idx], array[i] = array[i], array[min_idx]
        i += 1
    return array
array = [7,5,9,0,3,1,6,2,4,8]
selection_sort(array)

# 모범 답안
array = [7,5,9,0,3,1,6,2,4,8]
for i in range(len(array)):
    min_index = i
    for j in range(i+1, len(array)):
        if array[min_index] > array[j]:
            min_index = j
    array[i], array[min_index] = array[min_index], array[i]

print(array)
```

```python
# 삽입정렬 (Insertion Sorting)

# 내 답안 -> 틀림 ㅠ

# 1번째 고정
# 2 j번째는 1 i 번째랑만 비교
# 3 j 번째는 2,1 번째랑 비교
# 4번째는 3,2,1 번째랑 비교

# 왜 맨 앞이 안바뀔까 ?! 

def insertion_sort(array):
    for i in range(1, len(array)): # 바뀌는 대상 후보 위치
        cur = i
        if cur == 1:
            if array[cur] <= array[0]:
                array[cur], array[0] = array[0], array[cur]
        for j in range(i-1,0,-1): # 비교 대상 위치
            if array[cur] <= array[j]:
                array[cur], array[j] = array[j], array[cur]
                cur = j
    return array

array = [7,5,9,0,3,1,6,2,4,8]    
insertion_sort(array)


# 모범답안
def insertion_sort(array):
    for i in range(1, len(array)):
        for j in range(i,0,-1):
            if array[j] < array[j-1]:
                array[j], array[j-1] = array[j-1], array[j] # 굳이 i,j를 다 써서 비교할 필요가 없구나!
            else: # 안바뀌면 (= 자기보다 작은 데이터 만나면) 해당 위치에서 멈춤
                break
    return array

array = [7,5,9,0,3,1,6,2,4,8]    
insertion_sort(array)
```

```python
# 퀵 정렬

# 내 답안 -> 오 풀었다!!!!

def quick_sort(array):
    print(array)
    # 길이가 1보다 작으면 그냥 array return
    if len(array) <= 1:
        return array
    else:
        # pivot 설정
        pivot_idx = 0
        pivot = array[pivot_idx]
        while 1:
            start = 1
            end = pivot_idx
            # 왼쪽에서부터 pivot보다 큰 값 찾고, 찾으면 멈추기
            for i in range(len(array)):
                if pivot < array[i]:
                    start = i
                    break
            # 오른쪽에서부터 pivot보다 작은 값 찾고, 찾으면 멈추기
            for j in range(len(array)-1,1,-1):
                if pivot > array[j]:
                    end = j
                    break
            print(start, end)
            # 왼쪽 탐색 index < 오른쪽 탐색 index -> swap
            if start < end:
                array[start], array[end] = array[end], array[start]
            # 왼쪽 탐색 index > 오른쪽 탐색 index -> pivot과 swap 후 종료
            else:
                print(array)
                print(pivot_idx, start, end)
                array[pivot_idx], array[end] = array[end], array[pivot_idx]
                pivot_idx = end
                break
        # 재귀 탐색 위해 array[:pivot] array[pivot+1:] 이 2가지 실행하여 return 받고, merge하기
        array[:pivot_idx] = quick_sort(array[:pivot_idx])
        array[pivot_idx+1:] = quick_sort(array[pivot_idx+1:])

    return array

array = [7,5,9,0,3,1,6,2,4,8]    
quick_sort(array)

# 모범답안
def quick_sort(array, start, end):
    if start >= end: # 원소가 1개인 경우
        return
    
    pivot = start
    left = start + 1
    right = end

    while (left <= right):
        # 왼쪽에서 pivot보다 큰 값 찾을 때까지 반복
        while (left <= end and array[left] <= array[pivot]):
            left +=1
        # 오른쪽에서 pivot보다 작은 값 찾을 때까지 반복
        while (right > start and array[right] >= array[pivot]):
            right -=1
        
        # 엇갈린 case
        if (left > right):
            array[right], array[pivot] = array[pivot], array[right]
        # 안 엇갈린 case    
        else:
            array[left], array[right] = array[right], array[left]
    quick_sort(array, start, right - 1)
    quick_sort(array, right + 1, end)

array = [7,5,9,0,3,1,6,2,4,8]    
quick_sort(array, 0, len(array)-1)
array



# chat GPT 답안 + 파이썬 장점 살린 방식
def quick_sort(array):
    if len(array) <= 1:
        return array

    pivot = array[0]
    left = [x for x in array[1:] if x <= pivot]
    right = [x for x in array[1:] if x > pivot]

    return quick_sort(left) + [pivot] + quick_sort(right)

# 테스트
array = [7,5,9,0,3,1,6,2,4,8]  
sorted_array = quick_sort(array)
print(sorted_array)
```

```python
# 계수 정렬 

# 내 답안 -> ez
def count_sort(array):
    max_int = max(array)
    count_array = [0] * (max_int+1)
    for idx in array:
        count_array[idx] +=1
    sorted_array = []
    for i, cnt in enumerate(count_array):
        sorted_array += [i] * cnt

    return sorted_array

array = [7,5,9,0,3,1,6,2,9,1,4,8,0,5,2]

count_sort(array)

# 모범답안

def count_sort(array):
    count= [0] * (max(array)+1)

    for i in range(len(array)):
        count[array[i]] +=1
    
    for i in range(len(count)):
        for j in range(count[i]):
            print(i, end =' ')

count_sort(array)
```

```python
# 두 배열 원소 교체 문제
# 제한된 K번 교체 시도 내에서 배열 A 내 원소 합 최대로 만들기 

# A의 최소값과 B의 최댓값 찾아서 가장 작은 것부터 교환하기!
# 최대 K번이기 때문에, 교환을 안할 수도 있음 -> A의 최소값보다 B의 최댓값이 작으면 교환 멈추기!


# 과정 

# 5 3
# 1 2 5 4 3 
# 5 5 6 6 5

# 0. input 받기
n, k = list(map(int, input().split()))
a = list(map(int, input().split()))
b = list(map(int, input().split()))

cnt = 0

while cnt < k: # cnt가 K가 도달하면 종료하기
    # 1. A,B Sorting 하기
    a.sort()
    b.sort()    
    # 2. A의 최소값과 B의 최댓값 비교
    if b[-1] >= a[0]: # 2-1 만약 B가 더 크면 교체 & cnt +=1 해주기
        a[0], b[-1] = b[-1], a[0]
        cnt +=1
    else: # 2-2 만약 A가 더 크다면 종료
        break
print(sum(a))

# 모범답안이랑 동일!
# 단, 차이는

# b.sort(reverse=True)
# 로 역순으로 배열해서, 동일 index로 비교하도록 함

n, k = list(map(int, input().split()))
a = list(map(int, input().split()))
b = list(map(int, input().split()))

a.sort()
b.sort(reverse=True)

for i in range(k):
    if a[i] < b[i]:
        a[i], b[i] = b[i], a[i]
    else:
        break
print(sum(a))
```
